```python
datapath='./data/xiaohuangji50w_fenciA.conv'
dialog=[]
with open(datapath,'r',encoding='utf-8')as data:
    for line in data:
        if line[0]=='M':
            dialog.append(line.replace('M','').strip().split('/'))
print(len(dialog))
dialog[-1]
       
```

```python
vocab=['<pad>','<go>', '<eos>', '<unknown>']
for line in dialog:
    for i in range(len(line)):
        if line[i] not in vocab:
            vocab.append(line[i])
print(vocab)    

```

```python
print(len(vocab))
word2id={}
id2word={}
trainingSamples=[]
```

```python
for i in range(len(vocab)):
    word2id[vocab[i]]=i
    id2word[i]=vocab[i]
    
print(id2word)
```

```python
def lookup(line):
    ids=[]
    for i in range(len(line)):
        ids.append(word2id[line[i]])
    return ids
    

for i in range(0,len(dialog),2):
    dialog_pair=[]
    dialog_pair.append(lookup(dialog[i]))
    dialog_pair.append(lookup(dialog[i+1]))
    trainingSamples.append(dialog_pair)
    
print(trainingSamples[:10])    
    
```

```python
import pickle
output = open('./data/xiaohuangji.pkl', 'wb')
dataset={'word2id':word2id,'id2word':id2word,'trainingSamples':trainingSamples}
pickle.dump(dataset,output)
```
